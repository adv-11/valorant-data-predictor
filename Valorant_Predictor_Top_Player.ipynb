```python
import requests
import pandas as pd
userName= ""
userId=""
userName = input('Enter Valorant Username : ')
userId = input('Enter Valorant ID (dont include #) : ')
```

```python
puuid_url = "https://api.henrikdev.xyz/valorant/v1/account/"+userName+"/"+userId
resp_puuid = requests.get(puuid_url)
data_puuid = resp_puuid.json()
resp_puuid.json()
```

```python
puuid = data_puuid['data']['puuid']

comp_data_api = ""
comp_data_api = "https://api.henrikdev.xyz/valorant/v1/by-puuid/lifetime/matches/eu/"+puuid+"?mode=competitive"
comp_data_api_2 = "https://api.henrikdev.xyz/valorant/v1/lifetime/matches/ap/"+userName+"/"+userId+"?mode=competitive"
print(comp_data_api)
```

```python
resp_comp = requests.get(comp_data_api_2)
comp_data = resp_comp.json()
```

```python
matches_data = comp_data.get('data', [])
df = pd.json_normalize(matches_data)
df
```

```python
columns_to_drop = ['meta.map.id', 'meta.id', 'meta.version','meta.started_at' , 'meta.season.id' , 'meta.season.short' , 'stats.character.id' , 'stats.puuid' , 'stats.tier']
df = df.drop(columns=columns_to_drop, errors='ignore')

def determine_win(row):
  if (row['stats.team'] == 'Red') and (row['teams.red'] > row['teams.blue'])  :
     return('won')

  if (row['stats.team'] == 'Blue') and (row['teams.red'] < row['teams.blue']) :
     return('won')

  else : return('lost')


df.to_csv( userName, index=False)

df['stats.win']=df.apply(determine_win , axis=1)
df

```

```python
# @title stats.score vs stats.kills

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='stats.score', y='stats.kills', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

def determine_color(row):
    return 'green' if row['stats.win'] == 'won' else 'red'

# Add a new column 'bar_color' based on stats.win
df['bar_color'] = df.apply(determine_color, axis=1)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
ax = sns.countplot(y='stats.character.name', hue='bar_color', data=df, palette={'red': 'red', 'green': 'green'})
ax.set(xlabel='Frequency', ylabel='Character')
plt.title('Character Frequency and Win/Loss')

plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder


# Select relevant columns
selected_columns = ['meta.map.name', 'stats.character.name', 'stats.win']

# Create a new DataFrame with selected columns
df_model = df[selected_columns]

# Encode categorical variables
le_map = LabelEncoder()
le_character = LabelEncoder()

df_model['map_encoded'] = le_map.fit_transform(df_model['meta.map.name'])
df_model['character_encoded'] = le_character.fit_transform(df_model['stats.character.name'])

# Split data into features and target variable
X = df_model[['map_encoded', 'character_encoded']]
y = df_model['stats.win']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Combine the test set and predictions


df_results = pd.DataFrame({'map': le_map.inverse_transform(X_test['map_encoded']),
                            'character': le_character.inverse_transform(X_test['character_encoded']),
                            'actual_win': y_test,
                            'predicted_win': y_pred})

df_results['actual_win'] = df_results['actual_win'].map({'lost': 0, 'won': 1})

# Calculate win rates for each character on each map
win_rate_df = df_results.groupby(['map', 'character']).agg({'actual_win': 'mean'}).reset_index()

# Display the results
print(win_rate_df)
```
